# Runge Kutta Method

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return 2 * x

def runge_kutta(x0, y0, h, x_range):
    x_values = [x0]
    y_values = [y0]

    while x0 < x_range:
        k1 = h * f(x0, y0)
        k2 = h * f(x0 + h/2, y0 + k1/2)
        k3 = h * f(x0 + h/2, y0 + k2/2)
        k4 = h * f(x0 + h, y0 + k3)

        y1 = y0 + (k1 + 2*k2 + 2*k3 + k4)/6
        x0 += h
        y0 += y1

        x_values.append(x0)
        y_values.append(y0)

    return x_values, y_values

# Set the initial conditions, step size, and x range
x0 = 0.0
y0 = 0.0
h = 0.1
x_range = 1.0

# Solve the ODE using the Runge-Kutta method
x_values, y_values = runge_kutta(x0, y0, h, x_range)

# Plot the solution
plt.plot(x_values, y_values, 'b-', label='Numerical Solution')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of dy/dx = 2x using Runge-Kutta method')
```

```python
def decrypt(ciphertext, keytext):
    letters = "ABCDEFG"
    decrypted = ""

    for i in range(len(ciphertext)):
        ciphertext_letter = ciphertext[i]
        key_letter = keytext[i]

        # Calculate the index of the decrypted letter
        ciphertext_index = letters.index(ciphertext_letter)
        key_index = letters.index(key_letter)
        decrypted_index = (ciphertext_index - key_index) % len(letters)

        # Get the decrypted letter and append it to the decrypted message
        decrypted_letter = letters[decrypted_index]
        decrypted += decrypted_letter

    return decrypted

    ciphertext = "SYPOMNHALYDMKVKJPUYUOBEDIJKSSGSVYROBERLPGPPWVYKOMINJETVVSIM"
    keytext = "MGASUSXDREASUITVPLIMROEXUDULKWLIGLEXJAPENDEKNOKJGJOUYRUYJIJELO"

    decrypted = decrypt(ciphertext, keytext)

    print(decrypted)
```

# Newton raphson method

```python
import timeit
import math

def f(x):
    return 2*x**3 - 3**x - 12*x

def df(x):
    return 6*x**2 - 6*x-12

x0 = 1
tolerance = 1e-6
max_iterations = 100

for i in range(max_iterations):
    fx = f(x0)
    dfx = df(x0)
    
    x1 = x0 - (fx / dfx)
    
    print("Iterations " + str(i) + ": x = " + str(x0) + " f(x) = " +    
          str(f(x0)))
    
    if abs(f(x1)) < tolerance:
        print(f"Root found at x = {x1:.6f}")
        print("Time taken is: ", timeit.timeit())
        break
    else:
        x0 = x1
```

# Secant Method

```python
import numpy as np
import timeit
def f(x):
    return x + 2*np.cos(x)
def df(x):
    return 1 - 2*np.sin(x)

x0 = 2
x1 = 3
tol = 1e-6
max_iter = 100

for i in range(max_iter):
    fx0 = f(x0)
    fx1= f(x1)
    
    x_next = x1 - (fx1 * (x1 - x0) / (fx1 - fx0))
#     print(f"Iteration number is: {i}")
    
    print("Iteration" +str(i)+":x="+str(x_next)+"f(x_next)="+str(f(x_next)))

    if abs(f(x0)) < tol:
        print(f"Root found at:{x_next:.3f}")
        print("Time taken is", timeit.timeit())

    else:
        x0 = x1
        x1 = x_next
```

# Bisection Method

# Odeint in ODE

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def dy_dt(y, t):
    return (1 - 2*t) * y**2

y0 = 1
tspan = np.linspace(2, 10, 100)
y_exact = 1 / (tspan**2 - tspan + 1) 
y = odeint(dy_dt, y0, tspan)

plt.plot(tspan, y_exact, label='Exact Solution')
plt.plot(tspan, y, label='Numerical Solution')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

def roulette_simulation(bankroll, wager, num_spins):
    balance = [bankroll]
    for _ in range(num_spins):
        if random.choice(['red', 'black']) == 'red':
            bankroll += wager
        else:
            bankroll -= wager 
        balance.append(bankroll)
    return balance
10000
wager =
bankroll =  100
num_spins = 1000

balance = roulette_simulation(bankroll, wager, num_spins)

plt.plot(balance)
plt.xlabel('Spin Number')
plt.ylabel('Balance')
plt.title('Roulette Simulation')
plt.show()
```

```python
import numpy as np

def gauss_jordan_elimination(A, b):
    augmented_matrix = np.concatenate((A, b), axis = 1)
    n = A.shape[0]

    for i in range(n):
        augmented_matrix[i, :] /= augmented_matrix[i, i]
        for j in range(n):
            if i != j:
                augmented_matrix[j, :] -= (
                    augmented_matrix[j, i] * augmented_matrix[i, :]
                )

    x = augmented_matrix[:, n:]
    return x
# A = np.array([[2, -1, 3],
#               [1, 3, -2],
#               [-3, 1, -1]])

# b = np.array([[7],
#               [3],
#               [-2]])

# solution = gauss_jordan_elimination(A, b)
# print("Solution:")
# print(solution)

```
