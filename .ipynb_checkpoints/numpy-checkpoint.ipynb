# Scientific Computing


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

# Set the random seed for reproducibility
# random.seed(42)

# Create empty lists for hours studied and exam scores
hours_studied = []
exam_scores = []

# Generate random values for hours studied and exam scores
for _ in range(30):
    hours_studied.append(random.uniform(0, 10))
    exam_scores.append(random.uniform(0, 100))

# Create a dataframe from the generated data
data = pd.DataFrame({'Hours Studied': hours_studied, 'Exam Score': exam_scores})

# Plot a scatter plot to visualize the data points
sns.scatterplot(data=data, x='Hours Studied', y='Exam Score')
plt.title('Relationship between Hours Studied and Exam Score')
plt.show()

# Plot a line graph to show the trend between hours studied and exam scores
sns.lineplot(data=data, x='Hours Studied', y='Exam Score')
plt.title('Trend between Hours Studied and Exam Score')
plt.show()
```

```python
import numpy as np
from scipy.linalg import solve

# Define the coefficients matrix and the constants vector
A = np.array([[1, 1, 1],
              [25, 30, 70],
              [70, 30, 25]])
b = np.array([10,320,455])

# Solve the system of linear equations
solution = solve(A, b)

# Extract the amounts of each solution
x = solution[0]
y = solution[1]
z = 10 - x - y

# Print the solution
print("Amount of 25% solution:", x)
print("Amount of 30% solution:", y)
print("Amount of 70% solution:", z)
```

```python
import numpy as np

# Define the augmented matrix
A = np.array([[2, 3, -4],
              [5, 4, 2],
              [-3, 2, 1]])
B = np.array([1,2,3])

# Solve the system of equations
solution = np.linalg.solve(A,B)

x= solution[0]
y = solution[1]
z = solution[2]

# Print the solution
print("Solution:")
print(f"x = {solution[0]}")
print(f"y = {solution[1]}")
print(f"z = {solution[2]}")
```

# Newton Raphson Method

```python
import timeit

def f(x):
    return 2*x**3 - 3**x - 12*x

def df(x):
    return 6*x**2 - 6*x-12

x0 = 1
tolerance = 1e-6
max_iterations = 100

for i in range(max_iterations):
    fx = f(x0)
    dfx = df(x0)
    
    x1 = x0 - (fx / dfx)
    
    print("Iterations " + str(i) + ": x = " + str(x0) + " f(x) = " +    
          str(f(x0)))
    
    if abs(f(x1)) < tolerance:
        print(f"Root found at x = {x1:.6f}")
        print("Time taken is: ", timeit.timeit())
        break
    else:
        x0 = x1
```

# Secant Method

```python
import timeit

def f(x):
    return 2*x**3 - 3**x - 12*x

x0 = 1
x1 = 2
tolerance = 1e-6
max_iterations = 100

for i in range(max_iterations):
    fx0 = f(x0)
    fx1 = f(x1)
    
    x_next = x1 - (fx1 * (x1 - x0)) / (fx1 - fx0)
    
    print("Iteration " + str(i) + ": x = " + str(x_next) + " f(x) = " + str(f(x_next)))
    
    if abs(f(x_next)) < tolerance:
        print(f"Root found at x = {x_next:.6f}")
        print("Time taken is: ", timeit.timeit())
        break
    else:
        x0 = x1
        x1 = x_next
```

# Bisection Method

```python
import math
import timeit

def f(x):
    return 2*x**3 - 3**x - 12*x

a = 2
b = -1
tolerance = 1e-6
max_iterations = 100
    
for i in range(max_iterations):
    c = (a + b) / 2
    # print(f"Iteration number is: {i}")
    print("Iterations " + str(i) + ": x = " + str(c) + " f(x) = " +    
          str(f(c)))
    
    if abs(f(c)) < tolerance:
        print(f"Root found at x = {c:6f}")
        print("Time tacken is: ", timeit.timeit())
        break
    elif f(c) * f(a) < 0:
        b = c
    else:
        a = c
```
